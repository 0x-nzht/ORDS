```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def targetfunc(x, y):
    return x**2 + 2*x*y + 2*y**2 - 6*x + 4*y - 1

X, Y = np.meshgrid(np.arange(-5., 5., 0.5),
                   np.arange(-5., 5., 0.5))
Z = targetfunc(X, Y)


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

plt.show()
```

```python
import sympy as sym
sym.init_printing()
(x, y) = sym.symbols("x, y")


targetFunc = x**2 + 2*x*y + 2*y**2 - 6*x + 4*y - 1

def AnalyzeCurry(Z):
    dx = sym.diff(Z, x)             # x
    dy = sym.diff(Z, y)             # y
    s = sym.solve([dx, dy], [x, y]) # dx=0, dy=0
    return dx, dy, s


def main():
    dx, dy, s = AnalyzeCurry(targetFunc)
    print("x:", dx)
    print("y:", dy)
    print("", s)
 
if __name__ == '__main__':
    main()
```
