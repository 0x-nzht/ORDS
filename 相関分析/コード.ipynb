サンプルデータ

```python
import pandas as pd
import urllib.request
from io import StringIO

q = "https://raw.githubusercontent.com/0x-nzht/ORDS/main/%E7%9B%B8%E9%96%A2%E5%88%86%E6%9E%90/qlist.csv"
s = "https://raw.githubusercontent.com/0x-nzht/ORDS/main/%E7%9B%B8%E9%96%A2%E5%88%86%E6%9E%90/survey.csv"

#csvを読み込む関数
def read_csv(q):
    print(q)
    res = urllib.request.urlopen(q)
    res = res.read().decode("utf-8")
    df = pd.read_csv(StringIO( res) )
    return df

def read_csv(s):
    print(s)
    res = urllib.request.urlopen(s)
    res = res.read().decode("utf-8")
    df = pd.read_csv(StringIO( res) )
    return df

#実行
qlist=read_csv(q)
survey=read_csv(s)
```

```python
# 基本モジュール
import numpy as np
import pandas as pd

# 可視化のためのモジュール
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr_mat = survey.corr(method='pearson')
round(corr_mat,2)
```

```python
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            #annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )

plt.figure(figsize=(10, 10), dpi=50)
```
